```python
import torch, numpy as np, random
from datasets import load_dataset
from transformers import T5ForConditionalGeneration, T5Tokenizer
from src.metrics import compute_rouge  # your file
from src.sft import run_sft            # your file
from src.ppo import run_ppo            # your (debuggable) PPO file
```

```python
CFG = {
    "model_name": "google/flan-t5-small",
    "dataset": "cnn_dailymail",                 # "cnn_dailymail" or "xsum"
    "train_n": 5000,                            # SFT train size
    "val_n":   1000,
    "eval_n":  1000,                            # for generation/eval
    "max_input_len": 512,
    "max_target_len": 128,

    "gen": { "num_beams": 4, "max_new_tokens": 128, "batch_size": 32 },

    "sft": { "epochs_max": 3, "lr": 5e-5, "batch_size": 4,
             "warmup_ratio": 0.1, "patience": 1 },

    # We'll try PPO starting from both the BASE and the best SFT checkpoint
    "ppo_from": ["base", "sft"],

    # PPO config (now includes length filter + remainder + debug toggles)
    "ppo": {
        "n_train": 5000,          # raw samples to draw before filtering
        "min_len": 64,            # keep shorter articles too (increases survivors)
        "max_len": 2000,          # cap so inputs don't explode
        "batch_size": 4,
        "epochs": 1,
        "steps_per_epoch": None,  # <- will be filled in below based on survivors
        "remainder": "skip",      # ("skip" = drop short last batch; "wrap" = recycle)
        "max_new_tokens": 64,
        "length_penalty": 0.0,
        "lr": 1e-5,
        "ppo_epochs": 4,
        "target_kl": 0.1,
        "seed": seed,
    },

    # SCST (placeholder; fine to keep as-is for now)
    "scst": { "n_train": 2000, "epochs": 1, "batch_size": 8, "samples_per_input": 4 },
}
```

```python
# Tokenized for SFT
from src.data import load_cnn_dm_tokenized
tok, train_tok, val_tok = load_cnn_dm_tokenized(
    n_train=CFG["train_n"], n_val=CFG["val_n"],
    tokenizer_name=CFG["model_name"],
    max_input_len=CFG["max_input_len"], max_target_len=CFG["max_target_len"],
)

# Raw eval slice (generation)
eval_raw = load_dataset("cnn_dailymail","3.0.0")["validation"].select(range(CFG["eval_n"]))
len(train_tok), len(val_tok), len(eval_raw)
```

```python
from src.sft import run_sft  # src.sft
from src.eval_utils import generate_summaries_batched, metrics_table

best_dir = "./checkpoints/flan_t5_sft_best"
best_rl  = -1.0
bad      = 0
history_all = []

for ep in range(1, CFG["sft"]["epochs_max"]+1):
    out_dir = f"./checkpoints/flan_t5_sft_e{ep}"
    model, hist = run_sft(
        train_dataset=train_tok, val_dataset=val_tok, tokenizer=tok,
        model_name=CFG["model_name"], batch_size=CFG["sft"]["batch_size"],
        num_epochs=1, lr=CFG["sft"]["lr"], warmup_ratio=CFG["sft"]["warmup_ratio"],
        output_dir=out_dir,
    )
    # Evaluate on fixed settings
    preds, refs = generate_summaries_batched(
        model, tok, eval_raw,
        batch_size=CFG["gen"]["batch_size"],
        max_input_len=CFG["max_input_len"],
        max_new_tokens=CFG["gen"]["max_new_tokens"],
        num_beams=CFG["gen"]["num_beams"], do_sample=False,
    )
    scores = metrics_table(preds, refs)
    rl = scores["rougeL"]
    tr_loss = hist["train_loss"][-1] if hist["train_loss"] else float("nan")
    vl_loss = hist["val_loss"][-1]   if hist["val_loss"]   else float("nan")
    history_all.append((ep, tr_loss, vl_loss, scores))

    print(f"[SFT] epoch={ep}  ROUGE-L={rl:.4f}  train_loss={tr_loss:.4f}  val_loss={vl_loss:.4f}")
    if rl > best_rl + 1e-4:
        # Promote to best
        import shutil, os
        if os.path.exists(best_dir): shutil.rmtree(best_dir)
        shutil.copytree(out_dir, best_dir)
        best_rl = rl
        bad = 0
    else:
        bad += 1
        if bad > CFG["sft"]["patience"]:
            print("[SFT] Early stop (no ROUGE-L gain).")
            break

history_all, best_dir
```

```python
def eval_model(model_or_path):
    t = T5Tokenizer.from_pretrained(model_or_path)
    m = T5ForConditionalGeneration.from_pretrained(model_or_path)
    preds, refs = generate_summaries_batched(
        m, t, eval_raw,
        batch_size=CFG["gen"]["batch_size"],
        max_input_len=CFG["max_input_len"],
        max_new_tokens=CFG["gen"]["max_new_tokens"],
        num_beams=CFG["gen"]["num_beams"], do_sample=False,
    )
    return metrics_table(preds, refs)

base_scores = eval_model(CFG["model_name"])
sft_scores  = eval_model(best_dir)
print("BASE:", {k: round(v,4) for k,v in base_scores.items() if k in ["rouge1","rouge2","rougeL","bleu","compression","repetition"]})
print("SFT :", {k: round(v,4) for k,v in sft_scores.items()  if k in ["rouge1","rouge2","rougeL","bleu","compression","repetition"]})
```
